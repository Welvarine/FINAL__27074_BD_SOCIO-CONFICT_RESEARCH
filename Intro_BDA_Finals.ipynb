```python
import pandas as pd

# Re-import the dataset with proper encoding
df = pd.read_csv("WDR2011.csv", encoding='ISO-8859-1')
```

```python
df_recent = df[df['Year'].between(2000, 2009)]
```

```python
# Show all column names that contain certain keywords
keywords = ['gdp', 'edu', 'health', 'population', 'law', 'stability']

for word in keywords:
    print(f"\n=== Columns containing '{word}' ===")
    print([col for col in df.columns if word.lower() in col.lower()])
```

```python
selected_columns = [
    'Country', 'Countrycode', 'Year', 'RegionA', 'RegionB',
    'Cwbattledeaths_PRIO_UCDP',
    'CWintensity_PRIO_UCDP',
    'CWtype_PRIO_UCDP',
    'OnesidedStatec_UCDP',
    'OnesidedNonstate_UCDP',
    'Ruleoflaw_WGI',
    'MileduFromUS_USAID'
]
```

```python
df_focus =df_recent[selected_columns]
df_focus.head()
```

```python
df_focus.isnull().sum()
```

```python
# Total number of rows
total_rows = len(df_focus)

# % of rows missing governance value
missing_governance_pct = df_focus['Ruleoflaw_WGI'].isnull().mean() * 100

# % of rows missing military aid value
missing_aid_pct = df_focus['MileduFromUS_USAID'].isnull().mean() * 100

print(f"Missing 'Ruleoflaw_WGI': {missing_governance_pct:.2f}%")
print(f"Missing 'MileduFromUS_USAID': {missing_aid_pct:.2f}%")
```

```python
# Keep rows that have battle deaths or war intensity
df_focus_clean = df_focus.dropna(subset=[
    'Cwbattledeaths_PRIO_UCDP',
    'CWintensity_PRIO_UCDP'
])
```

```python
df_focus_clean['Ruleoflaw_WGI'] = df_focus_clean['Ruleoflaw_WGI'].fillna(0)
df_focus_clean['MileduFromUS_USAID'] = df_focus_clean['MileduFromUS_USAID'].fillna(0)
```

```python
df_focus_clean['Missing_RuleofLaw'] = df_focus['Ruleoflaw_WGI'].isnull().astype(int)
df_focus_clean['Missing_MilitaryAid'] = df_focus['MileduFromUS_USAID'].isnull().astype(int)
```

```python
df_focus_clean.isnull().sum()
```

```python
df_focus_clean['OnesidedStatec_UCDP'] = df_focus_clean['OnesidedStatec_UCDP'].fillna(0)
df_focus_clean['OnesidedNonstate_UCDP'] = df_focus_clean['OnesidedNonstate_UCDP'].fillna(0)
```

```python
df_focus_clean.isnull().sum()
```

```python
df_focus_clean.to_csv("cleaned_WDR2011_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set(style='darkgrid')
plt.rcParams['figure.figsize']=(10, 6)
```

```python
#GLOBAL BATTLE DEATHS over the years
sns.lineplot(data=df_focus_clean, x='Year', y='Cwbattledeaths_PRIO_UCDP', estimator='sum', ci=None)
plt.title('Total Global Battle Deaths per Year')
plt.ylabel('Total Battle Deaths')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create plot
sns.barplot(
    data=df_focus_clean,
    x='RegionB',
    y='Cwbattledeaths_PRIO_UCDP',
    estimator='sum',
    edgecolor='none'  # Remove bar outlines
)

# Title and labels
plt.title('Battle Deaths by Region (2000â€“2009)')
plt.ylabel('Total Deaths')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.grid(True)  # Optional: Add grid lines
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    data=df_focus_clean,
    x='Ruleoflaw_WGI',
    y='CWintensity_PRIO_UCDP',
    hue='RegionB',         # Color by region
    alpha=0.7              # Slight transparency for visibility
)

plt.title('Rule of Law vs. Conflict Intensity')
plt.xlabel('Rule of Law Score (WGI)')
plt.ylabel('Conflict Intensity (0 = none, 1 = minor, 2 = full-scale war)')
plt.grid(True)
plt.show()
```

```python
sns.boxplot(
    data=df_focus_clean,
    x='CWintensity_PRIO_UCDP',
    y='MileduFromUS_USAID',
    palette='viridis'
)
plt.title('Conflict Intensity vs Military Education Aid')
plt.xlabel('Conflict Intensity')
plt.ylabel('US Military Education Aid (in millions)')
plt.grid(True)
plt.show()
```

```python

avg_intensity = df_focus_clean.groupby('Country')['CWintensity_PRIO_UCDP'].mean().sort_values(ascending=False).head(20)

# Plot top 20 highest
sns.barplot(x=avg_intensity.values, y=avg_intensity.index, palette="rocket")
plt.title('Top 20 Countries by Average Conflict Intensity')
plt.xlabel('Average Conflict Intensity (0â€“2)')
plt.ylabel('Country')
plt.grid(True)
plt.show()
```

```python
# Filter countries with high conflict & weak governance
high_risk = df_focus_clean[
    (df_focus_clean['CWintensity_PRIO_UCDP'] >= 1.5) & 
    (df_focus_clean['Ruleoflaw_WGI'] < 0)
]

# Group by country and count how often they appear
targets = high_risk['Country'].value_counts().head(10)
print(targets)
```

```python
# Compare military aid across high vs low conflict countries
sns.boxplot(
    data=df_focus_clean,
    x='CWintensity_PRIO_UCDP',
    y='MileduFromUS_USAID',
    palette='Set1'
)

plt.title('Aid Distribution Across Conflict Intensity')
plt.xlabel('Conflict Intensity (0 = None, 2 = War)')
plt.ylabel('Military Education Aid (USD)')
plt.grid(True)
plt.show()
```
